```python
x=df.drop('income',axis=1)
y=df['income']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Ensure target is numeric (0/1). If not, map manually:
# y_train = y_train.map({'<=50':0, '>50':1})  # example mapping

# 1. Compute correlation of each feature with target
correlations = X_train.corrwith(y_train)

# 2. Sort by absolute correlation strength
correlations_sorted = correlations.abs().sort_values(ascending=False)

# 3. Select Top 15 features
top_15_features = correlations_sorted.head(15).index.tolist()

print("Top 15 features most correlated with target:")
print(top_15_features)

# Optional: see correlation values
print("\nCorrelation values:")
print(correlations[correlations_sorted.index[:15]])
```

```python
import matplotlib.pyplot as plt

correlations_sorted.head(15).plot(kind='barh', figsize=(8,6))
plt.title("Top 15 Features Correlated with Salary")
plt.xlabel("Correlation Strength (absolute)")
plt.gca().invert_yaxis()
plt.show()
```

```python

from imblearn.over_sampling import SMOTE
  # example model



# 2. Feature Selection (on original training set, imbalanced)
# suppose you already got top 10 features in prediction_features
prediction_features =correlations_sorted.index[:15]

X_train = X_train[prediction_features]
X_test = X_test[prediction_features]

# 3. Apply SMOTE only on training data
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
from imblearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

pipe_dt = Pipeline([
    ("smote", SMOTE(random_state=42)),
    ("clf", DecisionTreeClassifier(random_state=42))
])

param_dist_dt = {
    "clf__max_depth": randint(3, 40),
    "clf__min_samples_split": randint(2, 50),
    "clf__min_samples_leaf": randint(1, 30),
    "clf__max_features": [None, "sqrt", "log2"],
    "clf__criterion": ["gini", "entropy"],
}

rs_dt = RandomizedSearchCV(pipe_dt, param_dist_dt, n_iter=50, cv=5,
                           scoring="accuracy", random_state=42, n_jobs=-1,error_score='raise')
rs_dt.fit(X_train, y_train)
print(rs_dt.best_params_, rs_dt.best_score_)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

pipe_knn = Pipeline([
    ("scale", StandardScaler()),
    ("smote", SMOTE(random_state=42)),
    ("clf", KNeighborsClassifier())
])

param_grid_knn = {
    "clf__n_neighbors": list(range(3, 6, 2)),
    "clf__weights": ["uniform", "distance"],
    "clf__metric": ["minkowski"],
    "clf__p": [1, 2],   # 1=Manhattan, 2=Euclidean
}

gs_knn = GridSearchCV(pipe_knn, param_grid_knn, cv=5, scoring="accuracy", n_jobs=-1)
gs_knn.fit(X_train, y_train)
print(gs_knn.best_params_, gs_knn.best_score_)
```

```python

from sklearn.ensemble import GradientBoostingClassifier
from imblearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
pipe_gb = Pipeline([
    ('gb', GradientBoostingClassifier(random_state=42))
])

params_gb = {
    'gb__n_estimators': [100, 150],
    'gb__learning_rate': [0.05,0.1],
    'gb__max_depth': [3, 4],
    'gb__min_samples_split': [2],
    'gb__min_samples_leaf': [1]
}

grid_gb = GridSearchCV(pipe_gb, param_grid=params_gb, cv=5, scoring='accuracy', n_jobs=-1)
grid_gb.fit(X_train, y_train)

print("Best Gradient Boosting Params:", grid_gb.best_params_)
print("Best CV Accuracy:", grid_gb.best_score_ * 100)
```

```python



from sklearn.ensemble import RandomForestClassifier

pipe_rf = Pipeline([
    ('rf', RandomForestClassifier(random_state=42))
])

params_rf = {
    'rf__n_estimators': [100],          # fix at 100 for speed
    'rf__max_depth': [10, None],        # shallow vs full depth
    'rf__min_samples_split': [2],       # default is fine
    'rf__min_samples_leaf': [1, 2],     # small variation
    'rf__max_features': ['sqrt']
}

grid_rf = GridSearchCV(pipe_rf, param_grid=params_rf, cv=5, scoring='accuracy', n_jobs=-1)
grid_rf.fit(X_train, y_train)

print("Best Random Forest Params:", grid_rf.best_params_)
print("Best CV Accuracy:", grid_rf.best_score_ * 100)
33
```

```python

# Utility function for reporting
from sklearn.metrics import accuracy_score

def report_results(models, X_train, y_train, X_test, y_test):
    results = {}
    for name, search in models.items():
        best_model = search.best_estimator_

        # Train and test predictions
        y_train_pred = best_model.predict(X_train)
        y_test_pred = best_model.predict(X_test)

        train_acc = accuracy_score(y_train, y_train_pred) * 100
        test_acc = accuracy_score(y_test, y_test_pred) * 100
        cv_acc = search.best_score_ * 100

        results[name] = {
            "Best Params": search.best_params_,
            "Train Accuracy": train_acc,
            "Test Accuracy": test_acc,
            "CV Accuracy": cv_acc
        }

        print(f"====== {name} ======")
        print(f"Best Params: {search.best_params_}")
        print(f"Train Accuracy: {train_acc:.2f}%")
        print(f"Test Accuracy: {test_acc:.2f}%")
        print(f"CV Accuracy: {cv_acc:.2f}%\n")

    # Pick the best model based on CV
    best_model_name, best_search = max(models.items(), key=lambda kv: kv[1].best_score_)
    best_model = best_search.best_estimator_

    print(f"Best Model: {best_model_name}")
    print(f"Test Accuracy of Best Model: {best_model.score(X_test, y_test)*100:.2f}%")

    return results, best_model


# Example usage:
models = {
    "DT": rs_dt,
    "KNN": gs_knn,
    "RF": grid_rf ,
    "XGB": grid_gb
}

results, best_model = report_results(models, X_train, y_train, X_test, y_test)
```
